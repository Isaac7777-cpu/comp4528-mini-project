```python
import numpy as np
import torch
from torchvision.utils import save_image
from torchvision.io import read_image
from torchvision.transforms import v2
import matplotlib.pyplot as plt

generated_img_path = 'generated_images/conditional/Conditional-Image-Generation-Example/generated_image_editing.png'
original_img_path = 'generated_images/conditional/Conditional-Image-Generation-Example/original_image_editing.png'
generated_img = read_image(generated_img_path).float() / 255.0                 # Convert to float tensor [0,1]
original_img = read_image(original_img_path).float() / 255.0                 # Convert to float tensor [0,1]

def to_np_image(img: torch.Tensor) -> np.ndarray:
    return img.permute(1, 2, 0).numpy()

plt.subplot(1, 2, 1)
plt.imshow(to_np_image(generated_img))
plt.axis('off')
plt.title("Generated Imae")
plt.subplot(1, 2, 2)
plt.imshow(to_np_image(original_img))
plt.title("Original Input Image")
plt.axis('off')
```
