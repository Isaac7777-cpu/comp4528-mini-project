```python
import pandas as pd
import matplotlib.pyplot as plt

file_list = [
    "./logs/val/var_training_v0.5.0_depth=8_train_size=1024_epoc=100_mean_acc.csv",
    "./logs/val/var_training_v0.5.0_depth=8_train_size=1024_epoc=100_mean_loss.csv",
    "./logs/val/var_training_v0.5.0_depth=8_train_size=1024_epoc=100_recon_acc.csv",
    "./logs/val/var_training_v0.5.0_depth=8_train_size=1024_epoc=100_recon_loss.csv",
    "./logs/val/var_training_v0.5.1_depth=10_train_size=1024_epoc=100_mean_acc.csv",
    "./logs/val/var_training_v0.5.1_depth=10_train_size=1024_epoc=100_mean_loss.csv",
    "./logs/val/var_training_v0.5.1_depth=10_train_size=1024_epoc=100_recon_acc.csv",
    "./logs/val/var_training_v0.5.1_depth=10_train_size=1024_epoc=100_recon_loss.csv",
]

# Read them into a list of DataFrames
dataframes = [pd.read_csv(fname) for fname in file_list]
```

```python
dataframes[0].head()
```

```python
# Extract 'a' and 'e'
df_a = dataframes[0]
df_e = dataframes[4]

# Plot
plt.figure(figsize=(5, 5))
plt.plot(df_a['Step'], df_a['Value'] / 256, label='Model v0.5.0', color="#f9ab00")    # Extra handling as the first run
                                                                    # has accumulated mean accuracy across batch
plt.plot(df_e['Step'], df_e['Value'], label='Model v0.5.1', color="#9334e6")

plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Mean Accuracy Against Epoch Trained')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("./metrics/val/val_accm")
plt.show()
```

```python
# Extract 'a' and 'e'
df_0 = dataframes[2]
df_1 = dataframes[6]

# Plot
plt.figure(figsize=(5, 5))
plt.plot(df_a['Step'], df_0['Value'] / 256, label='Model v0.5.0', color="#f9ab00")    # Extra handling as the first run
# has accumulated mean accuracy across batch
plt.plot(df_e['Step'], df_1['Value'], label='Model v0.5.1', color="#9334e6")

plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Reconstruction (Tail) Accuracy Against Epoch Trained')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("./metrics/val/val_acct")
plt.show()
```
